```python
class HashDict:
    """
    真正的哈希函数和标准化的接口设计。
    """

    def __init__(self, capacity: int = 11):
        """
        初始化哈希表。
        
        Args:
            capacity: 哈希表容量
        """
        self.capacity = capacity
        self.table : list[list[tuple[str, int]]] = [[] for _ in range(capacity)]

    def _hash(self, key: str) -> int:
        """
        计算键的哈希值并映射到桶索引。
        
        Args:
            key: 要哈希的字符串键
            
        Returns:
            桶的索引（0 到 capacity-1）
        """
        return abs(hash(key)) % self.capacity

    def set(self, key: str, value: int) -> None:
        """
        插入或更新键值对。
        
        Args:
            key: 键（字符串）
            value: 值（整数）
        """
        bucket_idx = self._hash(key)
        bucket = self.table[bucket_idx]
        
        # 检查键是否已存在
        for i, (k, v) in enumerate(bucket):
            if k == key:
                # 更新现有键的值
                bucket[i] = (key, value)
                return
        
        # 键不存在，添加新键值对
        bucket.append((key, value))

    def get(self, key: str) -> int:
        """
        获取键对应的值。
        
        Args:
            key: 要查找的键
            
        Returns:
            键对应的值
            
        Raises:
            KeyError: 如果键不存在
        """
        bucket_idx = self._hash(key)
        bucket = self.table[bucket_idx]
        
        for k, v in bucket:
            if k == key:
                return v
                
        # 键不存在，抛出标准异常
        raise KeyError(f"Key '{key}' not found in the hash table.")

    def remove(self, key: str) -> bool:
        """
        删除指定的键值对。
        
        Args:
            key: 要删除的键
            
        Returns:
            True如果删除成功，False如果键不存在
        """
        bucket_idx = self._hash(key)
        bucket = self.table[bucket_idx]
        
        for i, (k, v) in enumerate(bucket):
            if k == key:
                # 删除找到的键值对
                del bucket[i]
                return True
                
        return False

    def contains(self, key: str) -> bool:
        """
        检查键是否存在于哈希表中。
        
        Args:
            key: 要检查的键
            
        Returns:
            True如果键存在，否则False
        """
        bucket_idx = self._hash(key)
        bucket = self.table[bucket_idx]
        
        for k, v in bucket:
            if k == key:
                return True
                
        return False

    def items(self):
        """返回哈希表中所有键值对的列表"""
        all_items = []
        for bucket in self.table:
            for item in bucket:
                all_items.append(item)
        return all_items

    def __str__(self) -> str:
        """
        返回哈希表的字符串表示。
        
        Returns:
            类似Python字典的字符串格式
        """
        result = []
        for bucket in self.table:
            for key, value in bucket:
                result.append(f"'{key}': {value}")
        
        return "{" + ", ".join(result) + "}"

    def clear(self) -> None:
        """清空哈希表"""
        self.table = [[] for _ in range(self.capacity)]
```

```python
if __name__ == "__main__":
    # 创建哈希表实例
    hd = HashDict()
    
    # 添加键值对
    hd.set("apple", 100)
    hd.set("banana", 200)
    hd.set("cherry", 300)

    # 查找
    print(hd.get("apple"))
    print(hd.get("banana"))

    # 更新
    hd.set("apple", 111)

    # 删除
    hd.remove("cherry")

    # 打印全部
    print(hd)
    print(hd.items())
    print(hd.__str__())

    # 清空哈希表
    print("\n清空哈希表:")
    hd.clear()
    print("清空后:", hd)
```
