```python
from collections.abc import Sequence
from typing import Any 
class Set:
    """
    自定义集合类，支持查询、插入、删除功能
    """
    def __init__(self,iterable:Sequence[Any]):
        """
        初始化集合

        参数:
        iterable: 可迭代对象，初始数据
        """
        if iterable is None:
            self.iterable = []
        else:
            self.iterable = list(iterable)
    def contains(self,item:Any) -> bool:
        """
        查询元素是否存在
        
        参数:
            item: 要查询的元素
            
        返回:
            bool: 是否存在
        """
        return item in self.iterable
    def insert(self,element:Any) -> bool:
        """
        插入元素（去重）
        
        参数:
            element: 要插入的元素
            
        返回:
            bool: 是否成功插入
        """
        if not self.contains(element):
            self.iterable.append(element)
            return True
        return False
    def remove(self,item:Any) -> bool:
        """
        删除元素
        
        参数:
            item: 要删除的元素
            
        返回:
            bool: 是否成功删除
            
        抛出:
            ValueError: 当元素不存在时
        """
        if  self.contains(item):
            self.iterable.remove(item)
            return True
        raise ValueError(f"元素 {item} 不存在")
```

```python
fruits=['s','a','c','d','p']
set1 = Set(fruits)
set1.insert('c')
set1.contains('a')
set1.remove('s')
print(set1.iterable)
```
