```python
from collections.abc import Sequence
from typing import Any

class OrderedArray:
    """
    插入有序数组类
    """
    def __init__(self, iterable: Sequence[Any]):
        """
        初始化有序数组
        
        参数:
            iterable: 可迭代对象，初始数据
        """
        if iterable is None:
            self.iterable = []
        else:
            self.iterable = list(iterable)
    def insert(self,element:Any) -> bool:
        """
        插入元素并保持有序
        
        参数:
            element: 要插入的元素
            
        返回:
            bool: 是否成功插入
        """
        insert_index = 0
        for index, value in enumerate(self.iterable):
            if element < value:
                insert_index = index
                break
        else:
            insert_index = len(self.iterable)
        self.iterable.insert(insert_index, element)    
```

```python
f = OrderedArray([0,11,22,33,44])
f.insert(25)
f.insert(50)
print(f.iterable)
```
